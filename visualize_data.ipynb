```python
import numpy as np
import pandas as pd
```

```python
loan_data = pd.read_csv('data/loan-approval-data.csv')
loan_data.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
loan_approval_by_gender = [
    len(loan_data[(loan_data['Gender']=='Male') & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Gender']=='Male') & (loan_data['Loan_Status']=='N')]),
    len(loan_data[(loan_data['Gender']=='Female') & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Gender']=='Female') & (loan_data['Loan_Status']=='N')])
]
loan_approval_by_gender_label = [
    f'Male Approved\n{loan_approval_by_gender[0]}',
    f'Male Not Approved\n{loan_approval_by_gender[1]}',
    f'Female Approved\n{loan_approval_by_gender[2]}',
    f'Female Not Approved\n{loan_approval_by_gender[3]}'
]
loan_approval_by_gender = np.array(loan_approval_by_gender)
plt.pie(loan_approval_by_gender, labels=loan_approval_by_gender_label, autopct='%1.1f%%')
plt.show()
```

```python
loan_approval_by_married = [
    len(loan_data[(loan_data['Married']=='Yes') & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Married']=='Yes') & (loan_data['Loan_Status']=='N')]),
    len(loan_data[(loan_data['Married']=='No') & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Married']=='No') & (loan_data['Loan_Status']=='N')])
]
loan_approval_by_married_label = [
    f'Married Approved\n{loan_approval_by_married[0]}',
    f'Married Not Approved\n{loan_approval_by_married[1]}',
    f'Not Married Approved\n{loan_approval_by_married[2]}',
    f'Not Married Not Approved\n{loan_approval_by_married[3]}'
]
loan_approval_by_gender = np.array(loan_approval_by_married)
plt.pie(loan_approval_by_married, labels=loan_approval_by_married_label, autopct='%1.1f%%')
plt.show()
```

```python
loan_approval_by_dependents = [
    len(loan_data[(loan_data['Dependents'] == 0) & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Dependents'] == 0) & (loan_data['Loan_Status']=='N')]),
    len(loan_data[(loan_data['Dependents'] == 1) & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Dependents'] == 1) & (loan_data['Loan_Status']=='N')]),
    len(loan_data[(loan_data['Dependents'] == 2) & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Dependents'] == 2) & (loan_data['Loan_Status']=='N')]),
    len(loan_data[(loan_data['Dependents'] == 3) & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Dependents'] == 3) & (loan_data['Loan_Status']=='N')]),
    loan_data[loan_data['Loan_Status']=='Y']['Dependents'].isna().sum(),
    loan_data[loan_data['Loan_Status']=='N']['Dependents'].isna().sum()
]
loan_approval_by_dependents_label = [
    f'0 Dependents Approved: {loan_approval_by_dependents[0]}',
    f'0 Dependents Not Approved: {loan_approval_by_dependents[1]}',
    f'1 Dependents Approved: {loan_approval_by_dependents[2]}',
    f'1 Dependents Not Approved: {loan_approval_by_dependents[3]}',
    f'2 Dependents Approved: {loan_approval_by_dependents[4]}',
    f'2 Dependents Not Approved: {loan_approval_by_dependents[5]}',
    f'3 Dependents Approved: {loan_approval_by_dependents[6]}',
    f'3 Dependents Not Approved: {loan_approval_by_dependents[7]}',
    f'nan Dependents Approved: {loan_approval_by_dependents[8]}',
    f'nan Dependents Not Approved: {loan_approval_by_dependents[9]}'
]
loan_approval_by_dependents = np.array(loan_approval_by_dependents)
plt.pie(loan_approval_by_dependents, labels=loan_approval_by_dependents_label, autopct='%1.1f%%')
plt.show()
```

```python
loan_approval_by_education = [
    len(loan_data[(loan_data['Education']=='Graduate') & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Education']=='Graduate') & (loan_data['Loan_Status']=='N')]),
    len(loan_data[(loan_data['Education']=='Not Graduate') & (loan_data['Loan_Status']=='Y')]),
    len(loan_data[(loan_data['Education']=='Not Graduate') & (loan_data['Loan_Status']=='N')])
]
loan_approval_by_education_label = [
    f'Graduate Approved\n{loan_approval_by_education[0]}',
    f'Graduate Not Approved\n{loan_approval_by_education[1]}',
    f'Not Graduate Approved\n{loan_approval_by_education[2]}',
    f'Not Graduate Not Approved\n{loan_approval_by_education[3]}'
]
loan_approval_by_education = np.array(loan_approval_by_education)
plt.pie(loan_approval_by_education, labels=loan_approval_by_education_label, autopct='%1.1f%%')
plt.show()
```
