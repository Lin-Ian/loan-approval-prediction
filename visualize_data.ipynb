```python
approved_loans_by_term = loan_data[['Loan_Amount_Term', 'Loan_Status']][loan_data['Loan_Status']=='Y'].groupby('Loan_Amount_Term').count().rename(columns={'Loan_Status': 'Approved'})
non_approved_loans_by_by_term = loan_data[['Loan_Amount_Term', 'Loan_Status']][loan_data['Loan_Status']=='N'].groupby('Loan_Amount_Term').count().rename(columns={'Loan_Status': 'Not Approved'})
loan_approval_by_term = pd.concat([approved_loans_by_term, non_approved_loans_by_by_term], axis=1)
loan_approval_by_term = loan_approval_by_term.sort_index()
loan_approval_by_term
```

```python
loan_approval_by_term.index = loan_approval_by_term.index.map(str)
loan_approval_by_term
```

```python
x_axis = np.arange(len(loan_approval_by_term.index))

plt.bar(x_axis-0.2, loan_approval_by_term['Approved'], 0.4, label='Approved')
plt.bar(x_axis+0.2, loan_approval_by_term['Not Approved'], 0.4, label='Not Approved')
plt.xticks(x_axis, loan_approval_by_term.index)
plt.xlabel('Loan Term')
plt.ylabel('Number of Applications')
plt.legend()
plt.show()
```

```python
x_axis = np.arange(len(loan_approval_by_term.index))

max_num_approved = loan_approval_by_term['Approved'].max()
max_num_not_approved = loan_approval_by_term['Not Approved'].max()

plt.bar(x_axis-0.2, loan_approval_by_term['Approved']/max_num_approved, 0.4, label='Approved')
plt.bar(x_axis+0.2, loan_approval_by_term['Not Approved']/max_num_not_approved, 0.4, label='Not Approved')
plt.xticks(x_axis, loan_approval_by_term.index)
plt.xlabel('Loan Term')
plt.ylabel('Number of Applications')
plt.legend()
plt.show()

# Impute median
```
