```python
import numpy as np
import pandas as pd
```

```python
loan_data = pd.read_csv('data/loan-approval-data.csv')
loan_data.head(10)
```

```python
loan_data.drop(['Loan_ID'], axis=1, inplace=True)
loan_data
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
loan_data['Gender'] = label_encoder.fit_transform(loan_data['Gender'])
loan_data
```

```python
loan_data['Married'] = label_encoder.fit_transform(loan_data['Married'])
loan_data
```

```python
loan_data['Dependents'] = loan_data['Dependents'].fillna(loan_data['Dependents'].median())
loan_data
```

```python
loan_data['Dependents'] = loan_data['Dependents'].astype(int)
loan_data
```

```python
loan_data['Education'] = label_encoder.fit_transform(loan_data['Education'])
loan_data
```

```python
loan_data['Self_Employed'] = label_encoder.fit_transform(loan_data['Self_Employed'])
loan_data
```

```python
loan_data['ApplicantIncome'] = loan_data['ApplicantIncome'].astype(float)
loan_data
```

```python
loan_data[['LoanAmount', 'Loan_Amount_Term']] = loan_data[['LoanAmount', 'Loan_Amount_Term']].fillna(loan_data[['LoanAmount', 'Loan_Amount_Term']].median())
loan_data
```

```python
loan_data['Credit_History'] = loan_data['Credit_History'].fillna(loan_data['Credit_History'].median())
loan_data
```

```python
loan_data['Credit_History'] = loan_data['Credit_History'].astype(int)
loan_data
```

```python
loan_data['Property_Area'] = label_encoder.fit_transform(loan_data['Property_Area'])
loan_data
```

```python
loan_data['Loan_Status'] = label_encoder.fit_transform(loan_data['Loan_Status'])
loan_data
```

```python
loan_data.to_csv('data/cleaned_data.csv', index=False)
```
